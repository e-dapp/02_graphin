Let's take all the important parts from the last post, make some lovely graphs from the data, and see if we can draw any conclusions from them.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

deaths = pd.read_csv('./data/deaths.csv')
deaths_no_flag = deaths.drop([col for col in deaths.columns if 'flag' in col], axis = 1)

areas = {}
unique_areas = pd.unique(deaths_no_flag['Reporting Area'].ravel())
for area in unique_areas:
    areas[area] = deaths_no_flag[deaths_no_flag['Reporting Area'] == area]

for area in unique_areas:
    areas[area] = areas[area].set_index(areas[area]['MMWR WEEK'])
for area in unique_areas:
    areas[area] = areas[area].drop([col for col in areas[area].columns if 'MMWR' in col], axis = 1)
```

That sums up all we accomplished last time. Doesn't look like a lot, but we'll be able to use these fundamentals to manipulate the data in other ways and to start visualizing this data. The only new thing I've added here is importing seaborn, which is a nifty visualization package that makes graphing and the like quite a bit easier. If I'm being totally honest, I like using it because seaborn's plots look much nicer than the corresponding plots from matplotlib without doing all kinds of customization. Before we do any new manipuation, let's get a little visalization of some of the data we were working with last time.

```python
plt.figure(figsize = (15, 10))
fig = sns.barplot(x = areas['New England'].index, 
            y = areas['New England']['All causes, by age (years), All Ages'], 
            hue = areas['New England']['Reporting Area'])
fig.set(xlabel = 'Week in Year', ylabel = 'Deaths - All Ages')
```

This is a fairly informative plot. We're looking at the deaths reported in New England by the week they were reported. We can easily find the min and max, the shape of the plot also seems to suggest a trend of more deaths happening on the edges of the year, it would be interesting to find if this trait is present in other regions. Let's write a little function to reproduce this graph for any Reporting Area we specify.

```python
def bar_by_week(area):
    plt.figure(figsize = (15, 10))
    fig = sns.barplot(x = areas[area].index, 
            y = areas[area]['All causes, by age (years), All Ages'], 
            hue = areas[area]['Reporting Area'])
    fig.set(xlabel = 'Week in Year', ylabel = 'Deaths - All Ages')
```

```python
bar_by_week('Boston, MA')
```

I'll just write me a quick bar plot function, to make things a little faster and easier.

```python
def quick_bar(d):
    '''Just a quick bar plot of all deaths by week in the year, d = week number'''
    
    plt.figure(figsize=(15, 10))
    fig = sns.barplot(x = weeks[d].index, y = weeks[d]['All causes, by age (years), All Ages'])
    fig.set(xlabel = 'index - can match to area', ylabel = 'Deaths - All Ages')
```

```python
quick_bar(1)
```

Let's augment this plotting function so we can choose how many entries, from the top, we'll include.

```python
def quick_bar2(d, e):
    '''Just a quick bar plot of all deaths by week in the year, d = week number
        plots the top e cities in the category'''
    
    plt.figure(figsize=(15, 10))
    fig = sns.barplot(x = weeks[d][:e].index, y = weeks[d][:e]['All causes, by age (years), All Ages'])
    fig.set(xlabel = 'index - can match to area', ylabel = 'Deaths - All Ages')
```

```python
quick_bar2(1, 10)
```

Aha, we can see this in the bigger plot a few cells up, but now we've got a good look at this. The first few entries are NaN values. We could just expand our number of entries, but let's be clean about this and get rid of the NaN's. NaN, by the way, are pandas way of working with NULL values, they are 'Not A Number.'

```python
for week in unique_weeks:
    weeks[week] = weeks[week][pd.notnull(weeks[week]['All causes, by age (years), All Ages'])]
```

```python
quick_bar2(1, 10)
```

132 unique areas, 1 of which we can chalk up to a total (which we've already taken care of), but that still leaves 9 areas which are showing up seemingly out of nowhere. Of course, they're not just random, they are actually sitting in our top 10. These extra areas are acting as subtotals, all of the areas that don't show up as cities are the total of the cities contained in those areas. Let's just collect them all and take them out of consideration. Then we'll only be dealing with cities, and that just keeps everything clean. Realizing this, we can go back later and evaluate these larger areas too.
